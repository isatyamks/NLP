```python
speech = """Good morning everyone,
It is an honor to be with you today. I want to speak about a topic that is very close to my heart: the power of dreams and how they shape the future of a nation.
India is a country rich in culture, resources, and above all, its youth. With over 65% of our population under the age of 35, we have the potential to be the world's leaders in science, technology, arts, and governance. But to achieve this, we must first dream.
Dreams are not what you see while sleeping; dreams are what keep you awake and motivated to achieve your goals. History is witness that every great accomplishment begins with a dream—a vision.
When we dreamed of freedom, we had leaders like Mahatma Gandhi who made it a reality. When we dreamed of self-reliance, we had scientists, engineers, and farmers who transformed our economy. Today, I urge each of you to dream of a stronger, self-sufficient, and united India.
Knowledge is the foundation of progress. It is your responsibility to constantly learn, innovate, and adapt. The world is changing rapidly with advancements in artificial intelligence, clean energy, and space exploration. We, as Indians, have the capability to lead in these fields.
When I worked on India's space program, we faced countless challenges. But every failure taught us something invaluable, and eventually, we succeeded. Remember, success does not come overnight—it demands hard work, perseverance, and teamwork."""
```

```python
words_list=nltk.word_tokenize(speech)
```

```python
words_list
```

```python
porter_stemmer = nltk.stem.PorterStemmer()
```

```python
porter_stemmer.stem(speech)
```

```python
new_list = words_list-stop_word
```

```python
set1 = set(words_list)
set2 = set(stop_word)
result = list(set1 - set2)
```

```python
result
```

```python
language = "english"
stemmer_Snowball = nltk.stem.SnowballStemmer(language)
```

```python
Snowball_words=[]
for word in result:
    Snowball_words.append(stemmer_Snowball.stem(word))
print(Snowball_words)
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer() 
lemmatization_words=[]
for word in result:
    lemmatization_words.append(lemmatizer.lemmatize(word,'english'))
print(lemmatization_words)
```
